```python
plt.figure(figsize=(70, 50), dpi=200)

# Filter the data to only include 2017 and 2018
df_filtered = df[df['shipping_year'].isin([2017, 2018])]

# Get the top 20 product categories
top_20_categories = df_filtered['product_category_grouped'].value_counts().index[:20]
df_filtered = df_filtered[df_filtered['product_category_grouped'].isin(top_20_categories)]

# Create the FacetGrid for breakdown by year and month
g = sns.FacetGrid(df_filtered, row="shipping_year", col="shipping_month", height=6, aspect=1, sharex=True, sharey=True)

# Map the countplot to the grid for product category counts
g.map(sns.countplot, 'product_category_grouped', order=top_20_categories)

# Adjust x-tick labels for better visibility in the FacetGrid
for ax in g.axes.flat:
    ax.set_xticks(range(len(top_20_categories)))  # Ensure fixed tick positions
    ax.set_xticklabels(top_20_categories, rotation=90, ha='right', fontsize=10)  # Set product category names

# Adjust layout for better visibility
plt.tight_layout()
plt.show()
```

```python

# Group by 'shipping_year' to count occurrences
yearly_counts = df.groupby('shipping_year').size().reset_index(name='count')

# Plot the time series for the Year
plt.figure(figsize=(12, 6))
sns.lineplot(x='shipping_year', y='count', data=yearly_counts)

# Labels and title
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Yearly Shipping Counts')

plt.tight_layout()
plt.show()  
```

```python


# Group by 'shipping_year' and 'shipping_month' to count occurrences
monthly_counts = df.groupby(['shipping_year', 'shipping_month']).size().reset_index(name='count')

# Create a datetime column to make it a time series (for plotting)
monthly_counts['date'] = pd.to_datetime(monthly_counts['shipping_year'].astype(str) + '-' + monthly_counts['shipping_month'].astype(str) + '-01')

# Plot the time series
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='count', data=monthly_counts)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Labels and title
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.title('Monthly Shipping Counts Over Time')

plt.tight_layout()
plt.show()
```
